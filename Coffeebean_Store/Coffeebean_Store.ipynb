# Coffeebean Store Crawling (2020-11-16 written)

```python
from bs4 import BeautifulSoup
import pandas as pd
import time
from selenium import webdriver

def CoffeebeanAddress(result):
    Coffeebean_URL = 'https://www.coffeebeankorea.com/store/store.asp'

    wd = webdriver.Chrome('C:/chromedriver.exe')
    wd.get(Coffeebean_URL)
    time.sleep(5)

    for idx in range(365):
        wd.get(Coffeebean_URL)
        wd.execute_script("storePop2(%s)" % str(idx+1))
        time.sleep(1)

        rcv_data = wd.page_source
        soupData = BeautifulSoup(rcv_data, 'html.parser')

        try:
            store_name = list(soupData.select('div.store_txt > h2'))[0].string

            store_info = soupData.select('div.store_txt > table > tbody > tr > td')
        
            store_hour = list(store_info)[0].string
            store_park = list(store_info)[1].string
            store_address = list(store_info[2])[0]
            store_number = list(store_info)[3].string

            print('[' + str(idx + 1) + '] ' + str(store_name) + ' is Successfully Crawled' )

            result.append([store_name, store_hour, store_park, store_address, store_number])

        except:
            print('[' + str(idx + 1) + '] is nowhere')
            result.append(['NA', 'NA', 'NA', 'NA', 'NA'])
            continue
        
        
        
    return

def cswin_Coffeebean():
    result = []

    print('Coffeebean CRAWLING START')
    CoffeebeanAddress(result)
    Coffeebean_table = pd.DataFrame(result, columns=('name', 'hour', 'park', 'address', 'number'))
    Coffeebean_table.to_csv("./Coffeebean.csv", encoding="cp949", mode='w', index=True)

    print('FINISHED')
    
    
if __name__ == '__main__':
     cswin_Coffeebean()
```
