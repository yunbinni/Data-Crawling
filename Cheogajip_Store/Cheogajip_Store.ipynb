```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import datetime
from itertools import count

import ssl  #접속보안 허용

def get_request_url(url, enc='utf-8'):
    
    req = urllib.request.Request(url)

    try:
        ssl._create_default_https_context = ssl._create_unverified_context    #접속보안 허용
        
        response = urllib.request.urlopen(req)
        if response.getcode() == 200:
            try:
                rcv = response.read()
                ret = rcv.decode(enc)
            except UnicodeDecodeError:
                ret = rcv.decode(enc, 'replace')    
            
            return ret
            
    except Exception as e:
        print(e)
        print("[%s] Error for URL : %s" % (datetime.datetime.now(), url))
        return None


def getCheogajipAddress(result):
    
    for page_idx in range(120):
        
        Cheogajip_URL = 'http://www.cheogajip.co.kr/bbs/board.php?bo_table=store&page=%s' % str(page_idx + 1)
        print ("[Cheogajip Page] : [%s]" % (str(page_idx + 1)))

        rcv_data = get_request_url(Cheogajip_URL)
        soupData = BeautifulSoup(rcv_data, 'html.parser')
           
        tbody = soupData.find('table').find('tbody')
        
        bEnd = True
        
        for store_tr in tbody.findAll('tr'):
            bEnd = False
            tr_tag = list(store_tr.strings)
            
            store_local = tr_tag[1]
            store_name = tr_tag[3]
            store_address = tr_tag[5]
            store_number = tr_tag[7]

            result.append([store_local] + [store_name] + [store_address] + [store_number])

        if (bEnd == True):
            print(result[0]) #확인용으로 출력
            print("== 데이터 수 : %d" %len(result))
            return

    return

def cswin_cheogajip():

    result = []
    print('CHEOGAJIP ADDRESS CRAWLING START')
    getCheogajipAddress(result)
    global pericana_table
    pericana_table = pd.DataFrame(result, columns=('local', 'name', 'address', 'number'))
    pericana_table.to_csv("./cheogajip.csv", encoding="cp949", mode='w', index=True)
    del result[:]
    print('FINISHED')
    
if __name__ == '__main__':
     cswin_cheogajip()

pericana_table.head()
```
